```python
# Importing Libraries
import seaborn as sns
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

Filter For Data Anaysts in United States...

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst' ) & (df['job_country'] == 'United States')]
df_DA_US
```

```python
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame() #converts it to data frame!

sns.set_theme(style='darkgrid')
sns.barplot(data = df_plot, x = 'count', y = 'job_location', hue = 'count', palette = 'YlOrBr_r', legend=False)
sns.despine()
plt.title("Number of Jobs per Job Titles")
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
df_plot = df['job_country'].value_counts().to_frame().head(15)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_country', hue='count', palette='YlOrBr_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Country')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
df_plot = df['company_name'].value_counts().to_frame()[1:].head(20)

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Company')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
